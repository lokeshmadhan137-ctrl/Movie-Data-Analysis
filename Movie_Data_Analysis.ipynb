```python
import pandas as pd
import random

# Base movie names
base_movies = [
    "Leo","Jailer","Avengers","Inception","Titanic","KGF","Vikram","RRR","Bahubali","Master"
]

genres = ["Action","Drama","Sci-Fi","Romance","Thriller","Comedy","Adventure"]

movies = []
ratings = []
votes = []
genre_list = []

for i in range(100):
    movie_name = random.choice(base_movies) + "_" + str(i+1)  # Unique names
    movies.append(movie_name)

    ratings.append(round(random.uniform(6.0, 9.5), 1))
    votes.append(random.randint(1000, 100000))
    genre_list.append(random.choice(genres))

data = pd.DataFrame({
    "movie": movies,
    "rating": ratings,
    "votes": votes,
    "genre": genre_list
})

data
```

```python
data.to_csv("movies_100.csv", index=False)
```
